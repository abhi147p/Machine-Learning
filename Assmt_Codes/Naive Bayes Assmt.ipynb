```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# function for reading the document
def read_documents(doc_file):
    docs, labels = [], []
    with open(doc_file, 'r', encoding='utf-8') as f:
        text_file = ""
        for line in f:
            text_file += line
        # removing the headers
        text_lists = text_file.split("\n")
        text_lists[9] = ' '.join(text_lists[9].split()[2:])
        all_docs = text_lists[9:len(text_lists)-1]
        # fetching the document words and labels
        for line in all_docs:
            words = line.strip().split()
            docs.append(words[3:])
            labels.append(words[1])
    return docs, labels
```

```python
# reading the document
all_docs, all_labels = read_documents('all_sentiment_shuffled.rtf')
```

```python
# getting the words count
words_map = {}
for doc in all_docs:
    for word in doc:
        if word in words_map:
            words_map[word] += 1
        else:
            words_map[word] = 1
word_count = list(words_map.values())
words = list(words_map.keys())
```

```python
# getting the labels of positive and negative and visualzing them

labels_count = {}
for label in all_labels:
    if label in labels_count:
        labels_count[label] += 1
    else:
        labels_count[label] = 1

pos_count, neg_count = labels_count['pos'], labels_count['neg']
plt.bar(range(len(labels_count)), list(labels_count.values()))
plt.xticks(range(len(labels_count)), list(labels_count.keys()))
plt.show()
```

```python
# splitting the data document into train and test data in 80-20

split_point = int(0.80*len(all_docs))
train_docs = all_docs[:split_point]
train_labels = all_labels[:split_point]  
eval_docs = all_docs[split_point:]
eval_labels = all_labels[split_point:]
```

```python
# finding unique words in a document

def find_unique(all_docs):
    unique_doc = []
    for doc in all_docs:
        for word in doc:
            if word not in unique_doc:
                unique_doc.append(word)
    return unique_doc

unique_words = find_unique(all_docs)
```

```python
len(unique_words)
```

```python
# writing the train_nb for computing the probabilities for each word. 
# i.e words in positive and negative sentences
# returns the list of positive and negative probabilities for each word.

def train_nb(documents, labels, unique_words):
    v = 0
    pos_tot_count = labels.count("pos")
    neg_tot_count = labels.count("neg")
    pos_prob, neg_prob = [], []
    for word in unique_words:
        posct, negct = 0,0
        for label_idx in range(len(labels)):
            for doc_word in documents[label_idx]:
                if word == doc_word and labels[label_idx] == "pos":
                    posct += 1
                elif word == doc_word and labels[label_idx] == "neg":
                    negct += 1
        
        # po = (posct/pos_tot_count)*(pos_tot_count/len(labels))
        po_1 = np.exp(np.log(posct) - np.log(pos_tot_count))
        po_2 = np.exp(np.log(pos_tot_count) - np.log(len(labels)))
        # po = po_1*po_2
        po = np.exp(np.log(po_1) + np.log(po_2))
        # no = (negct/neg_tot_count)*(neg_tot_count/len(labels))
        no_1 = np.exp(np.log(negct) - np.log(neg_tot_count))
        no_2 = np.exp(np.log(neg_tot_count) - np.log(len(labels)))
        # no = no_1*no_2
        no = np.exp(np.log(no_1) + np.log(no_2))
        # p = po/(po+no)
        p = np.exp(np.log(po) - np.log(po+no))
        # n = no/(no+po)
        n = np.exp(np.log(po) - np.log(po+no))
        pos_prob.append(p)
        neg_prob.append(n)
        v += 1
        if v%10 == 0:
            print(v)
    return pos_prob, neg_prob
```

```python
pos_prob, neg_prob = train_nb(all_docs,all_labels,unique_words)
```

```python
len(pos_prob), len(neg_prob)
```

```python
# function to evaluate the document data
def predict(all_docs, pos_prob, neg_prob, uniq_words):
    
    pred=[]
    v = 0
    for doc in all_docs:
        pos, neg = 0,0
        for word in doc:
            if word in uniq_words:
                ind = unique_words.index(word)
            else:
                continue
            pos+=pos_prob[ind]
            neg+=neg_prob[ind]
            
        if pos>neg:
            pred.append("pos")
        else:
            pred.append("neg")
        v += 1
#         if v%100 == 0:
#             print(v)
        
    return pred
```

```python
pred_labels = predict(eval_docs, pos_prob,neg_prob, unique_words)
```

```python
# function called accuracy_score to findout the accuracy of the prediction

def accuracy_score(pred, actual):
    r = 0
    for v in range(len(pred)):
        if pred[v]==actual[v]:
            r+=1
        else:
            continue
    return r/len(pred) * (100)
```

```python
print("Accuracy of testing data :",round(accuracy_score(pred_labels,eval_labels),2),"%")
```

```python
# evaluating N Fold cross validation

N = 7

# performs cross validation i,e created 7 folds and finds out the accuracy of each fold.

predicted=[]
for fold_nbr in range(N):
    split_point_1 = int(float(fold_nbr)/N*len(all_docs))
    split_point_2 = int(float(fold_nbr+1)/N*len(all_docs))

    train_docs_fold = all_docs[:split_point_1] + all_docs[split_point_2:]
    train_labels_fold = all_labels[:split_point_1] + all_labels[split_point_2:]
    eval_docs_fold = all_docs[split_point_1:split_point_2]
    eval_labels_fold = all_labels[split_point_1:split_point_2]
    predicted_labels = predict(eval_docs_fold,pos_prob, neg_prob,unique_words)
    predicted.append(predicted_labels)
    print(f"accuracy of {fold_nbr+1} Fold is :",round(accuracy_score(predicted_labels,eval_labels_fold),2),"%")
```
